```python
plt.figure(figsize=(6,4))

sns.countplot(x='Survived', data=df, hue='Survived', palette='Set2', legend=False)

plt.title("Survival Count of Titanic Passengers")
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")

plt.show()
```

### Observation

- Most passengers did not survive.
- The number of passengers who died is greater than the number of passengers who survived.
- The dataset is slightly imbalanced because the two classes do not have equal counts.

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Pclass', data=df, hue='Pclass', palette='Set2', legend=False)

plt.title("Passenger Class Distribution")

plt.xlabel("Passenger Class")

plt.ylabel("Number of Passengers")

plt.show()
```

### Observation

- The dataset contains more male passengers than female passengers.
- Male passengers account for the majority of the dataset.

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Pclass', data=df, hue='Pclass', palette='Set2', legend=False)

plt.title("Passenger Class Distribution")

plt.xlabel("Passenger Class")

plt.ylabel("Number of Passengers")

plt.show()
```

### Observation

- Most passengers traveled in Third Class.
- First Class had fewer passengers than Third Class.
- Second Class had the lowest number of passengers.

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=30, kde=True, color="skyblue")

plt.title("Age Distribution of Titanic Passengers")

plt.xlabel("Age")

plt.ylabel("Number of Passengers")

plt.show()
```

### Observation

- Most passengers were between 20 and 40 years old.
- Very few passengers were children or elderly.
- The age distribution is concentrated around young adults.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Age"], color="lightgreen")

plt.title("Box Plot of Passenger Age")

plt.xlabel("Age")

plt.show()
```

### Observation

- The Age column contains a few outliers.
- Most passengers fall within the middle age range.
- A few passengers are much younger or older than the majority.

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='Sex',
    hue='Survived',
    data=df,
    palette='Set2'
)

plt.title("Survival by Gender")

plt.xlabel("Gender")

plt.ylabel("Number of Passengers")

plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()
```

### Observation

- Female passengers had a higher survival rate than male passengers.
- Most male passengers did not survive.
- Although there were fewer female passengers, a larger proportion of them survived.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Pclass',
    hue='Survived',
    data=df,
    palette='Set2'
)

plt.title("Survival by Passenger Class")

plt.xlabel("Passenger Class")

plt.ylabel("Number of Passengers")

plt.legend(title="Survived", labels=["No", "Yes"])

plt.show()
```

### Observation

- First Class passengers had the highest survival rate.
- Third Class passengers had the lowest survival rate.
- Passenger class appears to have a strong relationship with survival.

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x="Age",
    hue="Survived",
    bins=30,
    kde=True,
    palette="Set2",
    multiple="layer"
)

plt.title("Age Distribution by Survival Status")

plt.xlabel("Age")

plt.ylabel("Number of Passengers")

plt.show()
```

### Observation

- Most passengers were between 20 and 40 years old.
- Children show relatively higher survival compared to some older age groups.
- Survival varies across different age ranges, indicating that age may have influenced survival.

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    x="Survived",
    y="Fare",
    data=df,
    palette="Set2"
)

plt.title("Fare Distribution by Survival Status")

plt.xlabel("Survived")
plt.ylabel("Fare")

plt.xticks([0,1],["No","Yes"])

plt.show()
```

### Observation

- Passengers who survived generally paid higher fares.
- Survivors have a higher median fare than non-survivors.
- The Fare column contains several outliers representing very expensive tickets.

```python
plt.figure(figsize=(10,8))

correlation = df.corr(numeric_only=True)

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

- Passenger Class and Fare show a moderate negative correlation.
- Fare has a weak positive correlation with Survival.
- Age has a weak correlation with Survival.
- Most variables have weak correlations, indicating that survival depends on multiple factors rather than a single feature.

```python
sns.pairplot(
    df,
    vars=["Age", "Fare", "Pclass", "SibSp", "Parch"],
    hue="Survived",
    palette="Set2"
)

plt.show()
```

### Observation

- Pair Plot provides relationships between multiple numerical variables.
- Fare and Passenger Class show a noticeable relationship.
- Most other numerical variables show weak relationships.
- Survival patterns can be observed through color differences across the plots.

# Summary of Findings

1. The dataset contains 891 passenger records and 12 columns before data cleaning.

2. Missing values were found in the Age, Cabin, and Embarked columns.
   - Age: 177 missing values
   - Cabin: 687 missing values
   - Embarked: 2 missing values

3. The Cabin column was removed because approximately 77% of its values were missing.

4. Missing values in the Age column were filled using the median, while missing values in the Embarked column were filled using the mode.

5. Most passengers did not survive the Titanic disaster.

6. The majority of passengers were male.

7. Most passengers traveled in Third Class.

8. The majority of passengers were between 20 and 40 years old.

9. Female passengers had a significantly higher survival rate than male passengers.

10. First Class passengers had a much higher survival rate than Third Class passengers.

11. Passengers who paid higher ticket fares generally had a better chance of survival.

12. The heatmap showed that Fare and Passenger Class have a moderate relationship, while most other numerical variables have weak correlations.

13. Pair Plot provided a comprehensive overview of relationships between numerical variables and survival.

# Conclusion

The Exploratory Data Analysis of the Titanic dataset provided valuable insights into passenger characteristics and survival patterns.

The analysis revealed that gender, passenger class, and ticket fare had a significant influence on survival. Female passengers and First Class passengers had considerably higher survival rates compared to male passengers and Third Class passengers. Most passengers were young adults, and age showed only a weak relationship with survival.

Data cleaning improved the quality of the dataset by handling missing values and removing the Cabin column due to excessive missing data. Various visualizations, including count plots, histograms, box plots, heatmaps, and pair plots, helped identify important trends, relationships, and patterns.

Overall, this EDA demonstrates how data visualization and statistical exploration can transform raw data into meaningful business insights, supporting better decision-making.
